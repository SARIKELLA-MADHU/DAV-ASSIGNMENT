```python
import pandas as pd
import numpy as np

# Load your dataset (assuming it's in memory from your provided data)
# For demonstration, I'll use a subset and split it
data = pd.read_csv('/content/weather_classification_data.csv')
```

```python
df1 = data.iloc[:50]  # First 50 rows
df2 = data.iloc[50:100]
print(df1)
print(df2)
```

```python
df3 = pd.DataFrame({
    'Location': ['inland', 'coastal', 'mountain'],
    'Region': ['Midwest', 'Coast', 'Highlands']
})
print("1. Concatenation (Vertical stacking):")
concat_df = pd.concat([df1, df2], axis=0, ignore_index=True)
print(concat_df.head())
```

```python
print("2. Append (using concat):")
append_df = pd.concat([df1, df2], axis=0, ignore_index=True)
print(append_df.head())
```

```python
print("3. Merge (inner join on Location):")
merged_df = pd.merge(concat_df, df3, on='Location', how='inner')
print(merged_df.head())
```

```python
print("Left Join:")
left_join = pd.merge(concat_df, df3, on='Location', how='left')
print(left_join.head())
```

```python
print("Right Join:")
right_join = pd.merge(concat_df, df3, on='Location', how='right')
print(right_join.head())
```

```python
print("Outer Join:")
outer_join = pd.merge(concat_df, df3, on='Location', how='outer')
print(outer_join.head())
```

```python
print("5. Aggregation and Grouping:")
grouped = concat_df.groupby(['Location', 'Season']).agg({
    'Temperature': ['mean', 'max', 'min'],
    'Humidity': ['mean'],
    'Precipitation (%)': ['sum']
})
print("Grouped statistics by Location and Season:")
print(grouped)
```

```python
print("6. Pivot Table:")
pivot_df = pd.pivot_table(
    concat_df,
    values='Temperature',
    index='Location',
    columns='Season',
    aggfunc='mean',
    fill_value=0
)
print("Temperature Pivot Table by Location and Season:")
print(pivot_df)
```

```python
print("7. Combined Operations Example:")
combined_df = pd.merge(concat_df, df3, on='Location', how='left')
pivot_combined = pd.pivot_table(
    combined_df,
    values='Temperature',
    index=['Region', 'Weather Type'],
    columns='Season',
    aggfunc='mean',
    fill_value=0
)
print("Pivot table with Region, Weather Type, and Temperature:")
print(pivot_combined)
```
